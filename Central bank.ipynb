# I will try and merge the dataset from central bank of Kenya

```
import os
#os.chdir(("drive//My Drive//Central_Bank_dataset"))
os.listdir()
```

```
import pandas as pd
remit = pd.read_csv("Diaspora Remittances.csv")
gdp  =  pd.read_csv('Annual GDP.csv')
exp = pd.read_csv('Central Government Expenditure  (Ksh Million).csv')
inf = pd.read_excel('Inflation Rates.xlsx')
foreign = pd.read_csv('Foreign Trade Summary (Ksh Million).csv')
exc = pd.read_csv('Key CBK Indicative Exchange Rates.csv')
mpesa = pd.read_csv('Mobile Payments.csv')
remit.columns = [i if i in ["Year","Month"] else i+"_Remittances" for i in remit.columns]
remit = remit[["Year","Month","Total Remittances (USD '000)_Remittances"]]
remit.columns = ["Year","Month","Total_Remittances(USD)"]
remit.head()
```

```

months = {'August':8, 'July':7,'June':6, 'May':5, 'April':4, 'March':3, 'February':2,
       'January':1, 'December':12, 'November':11, 'October':10, 'September':9}
print(months.get('October',"Unknown"))
exp["Month"] =[months.get(i,"Unknown") for i in  exp["Expenditures Month"]]
exp = exp[["Total Recurrent Expenditure","Development Expenditure","Total Expenditure","Year","Month"]]
exp.head()
```

```
final = pd.merge(exp,remit,on = ["Year","Month"],how = "left")
final.head(),final.shape
```

```
inf["Month"] = [months.get(i,"Unknown") for i in inf.Month]
inf= inf[['Year', 'Month', 'Annual Average Inflation']]
inf.columns = ['Year', 'Month', 'Inflation']
final = pd.merge(final,inf,on = ["Year","Month"],how = "left")
final.head(),final.shape
```

```
foreign1 = foreign[["Year","Month","Trade Balance","Total"]].copy()
foreign1["Month"] = [months.get(i,"unknown") for i in foreign.Month]
foreign1.columns = ["Year","Month","Trade_Balance","Total_Imports"]
final = pd.merge(final.copy(),foreign1.copy(),on = ["Year","Month"],how = "left")
final.head(),final.shape
final.head()
```

```
final1 = pd.merge(final,gdp,on ="Year",how = "left")
print(final1.columns)
final1.head()
```

```
final1 =final1[['Year', 'Month','Total Recurrent Expenditure', 'Development Expenditure',
       'Total Expenditure', 'Total_Remittances(USD)',
       'Inflation', 'Trade_Balance', 'Total_Imports',
       'Nominal GDP prices (Ksh Million)', 'Annual GDP growth (%)',
       'Real GDP prices (Ksh Million)']]


final1.head()
```

```
final_df = final1.sort_values(["Year","Month"],ascending=False).copy()
print(final_df.shape)
final_df.head()
```

```
final_df.to_csv("final_merged_data.csv")
```
